```python
# Importing libraries
import pandas as pd
import numpy as np
```

```python
# Loading the dataset
df = pd.read_csv("googleplaystore(impure).csv")
df.head()
```

```python
# Getting information about the dataset
df.info()
```

```python
# Checking the null values
df.isnull().sum()
```

- Here "Unnamed: 0" is an unnecessary column

```python
# Checking for duplicate records
df.duplicated().sum()
```

- Now we are going to do the cleaning operation

```python
# Cleaning part
def clean(df):
    return(
        df
        .dropna()
        .rename(columns=str.lower)
        .rename(columns={'content rating':'content_rating','last updated':'last_updated','current ver':'current_ver','android ver':'android_ver'})
        .assign(
            last_updated = lambda df_ : pd.to_datetime(df_['last_updated'],errors='coerce').dt.strftime('%Y-%m-%d')
        )
        .assign(last_updated =lambda df_ : pd.to_datetime(df_['last_updated'],errors='coerce'))
        .drop('unnamed: 0',axis=1)
    )

cleaned_df = clean(df)
cleaned_df
```

```python
cleaned_df.info()
```
