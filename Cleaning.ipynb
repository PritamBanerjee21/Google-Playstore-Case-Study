- Now we are going to do the cleaning operation

```python
# Cleaning part
def clean(df):
    return(
        df
        .drop(index=10472,axis=0)
        .dropna()
        .rename(columns=str.lower)
        .rename(columns={'content rating':'content_rating','last updated':'last_updated','current ver':'current_ver','android ver':'android_ver'})
        .assign(
            last_updated = lambda df_ : pd.to_datetime(df_['last_updated'],errors='coerce').dt.strftime('%Y-%m-%d')
            )
        .assign(last_updated =lambda df_ : pd.to_datetime(df_['last_updated'],errors='coerce'))
        .assign(installs = lambda df_ : (
            df_.installs
            .str.replace('+','')
            .str.replace(',','')
            .astype(int))
               )
        .assign(price = lambda df_ : (
            df_.price
            .str.replace('$','')
            .astype(float)
        ))
        .assign(reviews = lambda df_ : df_.reviews.astype(int))
        .drop('unnamed: 0',axis=1)

    )

cleaned_df = clean(df)
cleaned_df
```
